# Exploratory Data Analysis (EDA) untuk OmniMeshV2

Notebook ini digunakan untuk eksplorasi data sebelum training.

## 1. Load Data

```python
import pandas as pd
import glob

data_dir = "../data/raw/"
files = glob.glob(data_dir + "*.csv")
df_list = []
for f in files:
    df = pd.read_csv(f)
    df_list.append(df)
    
if df_list:
    df = pd.concat(df_list, ignore_index=True)
    print(f"Total samples: {len(df)}")
    print(df.head())
```

## 2. Distribusi Label

```python
if 'label' in df.columns:
    print(df['label'].value_counts())
    df['label'].value_counts().plot(kind='bar')
```

## 3. Panjang Teks

```python
if 'text' in df.columns:
    df['text_length'] = df['text'].astype(str).apply(len)
    print(df['text_length'].describe())
```
